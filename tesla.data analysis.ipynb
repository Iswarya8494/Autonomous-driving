```python
filtered_df = part2_data.dropna(how = 'all')
print(filtered_df)
```

```python
df_duplicated = filtered_df[filtered_df.duplicated()]
print(df_duplicated)
```

```python
print(filtered_df.columns.tolist())
```

```python
unamed_columns = [col for col in filtered_df if 'Unnamed' in col]
filtered_df = filtered_df.drop(columns = unamed_columns, axis = 1)
print(filtered_df)
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns

filtered_df.dropna(subset=['Date',' State ',' Country ','Year'], inplace=True)

event_by_date = filtered_df.groupby('Date').size()
event_by_year = filtered_df.groupby('Year').size()
event_by_country = filtered_df.groupby(' Country ').size()
event_by_state = filtered_df.groupby(' State ').size()

plt.figure(figsize = (10,6))
event_by_date.plot(kind='bar')
plt.title('Number of Events by date')
plt.xlabel('date')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize = (10,6))
event_by_year.plot(kind='bar')
plt.title('Number of Events by year')
plt.xlabel('year')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize = (10,6))
event_by_state.plot(kind='bar')
plt.title('Number of Events by state')
plt.xlabel('state')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize = (10,6))
event_by_country.plot(kind='bar')
plt.title('Number of Events by country')
plt.xlabel('country')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()
```

```python
accidents_deaths = filtered_df.groupby('Case #')[' Deaths '].sum()
print(accidents_deaths.max())
print(accidents_deaths.min())
accidents_deaths
```

```python
filtered_df[' Tesla driver '] = pd.to_numeric(filtered_df[' Tesla driver '], errors='coerce')
filtered_df[' Tesla driver '].fillna(0, inplace=True)
times_tesla_drivers_die = filtered_df[' Tesla driver '].sum()
times_tesla_drivers_die
```

```python
 ' Tesla occupant '
filtered_df[' Tesla occupant '] = pd.to_numeric(filtered_df[' Tesla occupant '], errors='coerce')
filtered_df[' Tesla occupant '].fillna(0, inplace=True)
times_tesla_occupants_die = (filtered_df[' Tesla occupant ']>0).sum()
Total_accidents = filtered_df['Case #'].nunique()
propotaion = times_tesla_occupants_die/Total_accidents
print(f"the propotion of tesla ocupangt cases are:{propotaion:.2%}")
print(f"total acciedents and ocupant dies:{Total_accidents} {times_tesla_occupants_die}")
```

```python
print(filtered_df[' Cyclists/ Peds '].head(10))
print (filtered_df[' Other vehicle '].head(10))
```

```python
filtered_df[' Cyclists/ Peds '] = pd.to_numeric(filtered_df[' Cyclists/ Peds '], errors='coerce')
filtered_df[' Cyclists/ Peds '].fillna(0, inplace=True)
times_cyclist_ped_die = (filtered_df[' Cyclists/ Peds ']>0).sum()
propotaion_cyc_ped = times_cyclist_ped_die/Total_accidents
print(f"the propotion of cyclist_ped cases are:{propotaion_cyc_ped:.2%}")
print(f"total acciedents and ocupant dies:{Total_accidents} {times_cyclist_ped_die}")
categories = ['With Cyclists/Peds', 'Without Cyclists/Peds']
counts = [times_cyclist_ped_die, Total_accidents-times_cyclist_ped_die]
plt.bar(categories, counts, color=['blue', 'green'])
plt.title('Distribution of Events Involving Cyclists or Pedestrians')
plt.ylabel('Number of Events')
plt.show()
```

```python
lst = 0
for index, row in filtered_df.iterrows():
    if row[' Cyclists/ Peds '] > 0:
        if row[' Tesla driver ']>0 or row[' Tesla occupant '] > 0:
            lst = lst + 1
print(f"The number of accidents involving Tesla deaths along with cyclists or pedestrians: {lst}")
        
```

```python
filtered_df[' Other vehicle '] = pd.to_numeric(filtered_df[' Other vehicle '],  errors='coerce')
filtered_df[' Other vehicle '].fillna(0, inplace = True)
frequency_other_vehicle_hit = (filtered_df[' Other vehicle ']>0).sum()
print(frequency_other_vehicle_hit)
```

```python
model_grouping = filtered_df.groupby(' Model ')['Case #'] .count()
categories = model_grouping.index  # This will give you the model categories
count = model_grouping.values 
plt.bar(categories, count)
plt.title("distribution of event wrt model")
plt.xlabel("Model")
plt.ylabel("Count of Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
filtered_df[' Verified Tesla Autopilot Deaths '].replace('-' , 0, inplace =True)
filtered_df[' Verified Tesla Autopilot Deaths '].fillna(0, inplace= True)
print(filtered_df[' Verified Tesla Autopilot Deaths '].unique())
verified_tesla_autopilot_deaths_grouping = filtered_df[' Verified Tesla Autopilot Deaths '].value_counts()
category = verified_tesla_autopilot_deaths_grouping.index
values = verified_tesla_autopilot_deaths_grouping.values
plt.bar(category, values)
plt.title("distribution of count of verified tesla autopilot deaths")
plt.xlabel("")
plt.ylabel("Count of Cases")
plt.xticks(rotation=45)
plt.show()
```
